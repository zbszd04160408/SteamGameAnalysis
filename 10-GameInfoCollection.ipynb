```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

headers = {
    'Accept': '',
    'Accept-Encoding': '',
    'Accept-Language': '',
    'Cache-Control': '',
    'Connection': '',
    'Cookie':'',
    'Host': '',
    'Sec-Fetch-Mode': '',
    'Sec-Fetch-Site': '',
    'Sec-Fetch-User': '',
    'Upgrade-Insecure-Requests': '',
    'User-Agent': ''
}

n = 501

path = '1.xlsx'

def getgamelist(n):
    linklist=[]
    IDlist = []
    for pagenum in range(1,n):
        r = requests.get('https://store.steampowered.com/search/?category1=998&filter=topsellers&page=%d'%pagenum,headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36'})
        soup = BeautifulSoup(r.text, 'lxml')
        soups= soup.find_all(href=re.compile(r"https://store.steampowered.com/app/"),class_="search_result_row ds_collapse_flag")
        for i in soups:
            i = i.attrs
            i = i['href']
            link = re.search('https://store.steampowered.com/app/(\d*?)/',i).group()
            ID = re.search('https://store.steampowered.com/app/(\d*?)/(.*?)/', i).group(1)
            linklist.append(link)
            IDlist.append(ID)
        print('Completed'+str(pagenum)+'pages, currently'+str(len(linklist)))
    return linklist,IDlist

def getdf(n):
    linklist,IDlist = getgamelist(n)
    df = pd.DataFrame(list(zip(linklist,IDlist)),
               columns =['Link', 'ID'])
    return df
if __name__ == "__main__":
    df = getdf(n)
    df.to_excel(path)
```

```python
# import requests
# cookies = {'birthtime': '568022401'}
# r = requests.get('http://store.steampowered.com/', cookies=cookies)

def gamename(soup):   #name
    try:
        a = soup.find(class_="apphub_AppName")
        k = str(a.string)
    except:
        a = soup.find(class_="apphub_AppName")
        k = str(a.text)
    return k

def gameprice(soup):#price
    try:
        a = soup.findAll(class_="discount_original_price")
        for i in a:
            if re.search('$|free', str(i),re.IGNORECASE):
                a = i
        k = str(a.string).replace('	', '').replace('\n', '').replace('\r', '').replace(' ', '')
    except:
        a = soup.findAll(class_="price")
        for i in a:
             if re.search('$|free', str(i),re.IGNORECASE):
                a = i
        k = str(a.string).replace('	', '').replace('\n', '').replace('\r', '').replace(' ', '')
    return k

def taglist(soup):#tags
    list1=[]
    a = soup.find_all(class_="app_tag")
    for i in a:
        k = str(i.string).replace('	', '').replace('\n', '').replace('\r', '')
        if k == '+':
            pass
        else:
            list1.append(k)
    list1 = str(','.join(list1))
    return list1

def description(soup):  #description
    a = soup.find(class_="game_description_snippet")
    k = str(a.string).replace('	', '').replace('\n', '').replace('\r', '')
    return k

def allreviewsummary(soup):   #All Reviews
    a = soup.find(class_="summary column")
    try:
        k = str(a.span.string)
    except:
        k=str(a.text)
    return k

def recentreviewsummary(soup):   #Recent Reviews
    a = soup.find(class_="summary_section")
    try:
        k = str(a.span.string)
    except:
        k=str(a.text)
    return k

def getdate(soup):   # Release date
    a = soup.find(class_="date")
    k = str(a.string)
    return k

def userreviewsrate(soup):#positve rate
    a = soup.find(class_="user_reviews_summary_row")
    k = str((a.attrs)['data-tooltip-html'])
    return k

def developer(soup):   #developer
    a = soup.find(id="developers_list")
    k = str(a.a.string)
    return k

# def getreviews(ID):# reviews
#     cookies = { 'birthtime': '-2211667760', 'lastagecheckage': '1-0-1900'}
#     r1 = requests.get(
#         'https://store.steampowered.com/appreviews/%s'%str(ID),headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36'},timeout=10, cookies = cookies)
#     soup = BeautifulSoup(r1.json()['html'], 'lxml')
#     a = soup.findAll(class_="content")
#     list1 = []
#     for i in a:
#         list1.append(i.text.replace('	', '').replace('\n', '').replace('\r', '').replace(' ', ','))
#     k=str('\n'.join(list1))
#     return k

def getdetail(x):
    
    tag, des, reviews, date, rate, dev, review,name,price = ' ', ' ', ' ', ' ', ' ', ' ', ' ',' ',' '
    global count
    try:
        cookies = { 'birthtime': '-2211667760', 'lastagecheckage': '1-0-1900'}
        r = requests.get(x['Link'], headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36'},timeout=10, cookies = cookies)
    except:
        print('No respond1')
        try:
            r = requests.get(x['Link'], headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36'},timeout=10)
        except:
            print('No respond2')
            try:
                r = requests.get(x['Link'], headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36'},timeout=10)
            except:
                print('No respond3')
        
    soup = BeautifulSoup(r.text, 'lxml')
    try:
        name = gamename(soup)
    except:
        name = "error"

    try:
        tag = taglist(soup)
    except:
        tag = "error"

    try:
        des = description(soup)
    except:
        des = "error"

    try:
        allreviews = allreviewsummary(soup)
    except:
        allreviews = "error"

    try:
        recentreviews = recentreviewsummary(soup)
    except:
        recentreviews = "error"
    
    try:
        date = getdate(soup)
    except:
        date = "error"
    
    try:
        rate = userreviewsrate(soup)
    except:
        rate = "error"
    
    try:
        dev = developer(soup)
    except:
        dev = "error"
        # review = getreviews(str(x['ID']))
    try:
        price = gameprice(soup)
    except:
        price = "error"
        
    print('Completed: '+name+str(x['ID'])+' %d: '%count)
    

    count += 1
    return name,price,tag,des,allreviews,recentreviews,date,rate,dev#,review



df1 = pd.read_excel('1.xlsx')
count = 1
df1['Details'] = df1.apply(lambda x: getdetail(x), axis=1)
df1['Name'] = df1.apply(lambda x: x['Details'][0], axis=1)
df1['Price'] = df1.apply(lambda x: x['Details'][1], axis=1)
df1['Tag'] = df1.apply(lambda x: x['Details'][2], axis=1)
df1['Description'] = df1.apply(lambda x: x['Details'][3], axis=1)
df1['All Reviews'] = df1.apply(lambda x: x['Details'][4], axis=1)
df1['Recent Reviews'] = df1.apply(lambda x: x['Details'][5], axis=1)
df1['Release Date'] = df1.apply(lambda x: x['Details'][6], axis=1)
df1['Positive Rate'] = df1.apply(lambda x: x['Details'][7], axis=1)
df1['Developer'] = df1.apply(lambda x: x['Details'][8], axis=1)
# df1['评论'] = df1.apply(lambda x: x['Details'][9], axis=1)

df1.to_csv('Game_Info.csv', index=False)
print('Completed')
```
